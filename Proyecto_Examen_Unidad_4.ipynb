```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Diana2004450/Unidad-4/refs/heads/main/dataset%20(1).csv")
df.drop(columns=[], inplace=True)
df.dropna(inplace=True)
df
```

# **participacin parental**

```python
df['participacion_parental'].unique()
```

```python
import scipy.stats as stats
nivel_de_significancia =0.05

#Realizar ANOVA
anova_result = stats.f_oneway(
      df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
      df[df['participacion_parental'] == 'medio']['puntaje_examen'],
      df[df['participacion_parental'] == 'elevado']['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos de los grupos.")
else:
    print("Conclusin: No hay diferencias significativas entre los grupos.")
```

El F-Estadstico de 80.46 indica que la variacin entre las medias de los grupos, es decir, los niveles de participacin parental es mucho mayor que la variacin dentro de los grupos. El valor p = 0.00000 es mucho menor que el nivel de significancia de 0.05, entonces confirma que hay diferencias significativas entre las medias de al menos dos niveles de participacin parental, obteniendo como conclusin que si existen diferencias significativas entre al menos dos de los grupos.

```python
# Si todos los grupos tienen una distribucin normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los grficos de Tukey para tomar una decisin (ignorando la tabla).
#.    Estrictamente hablando, tendran que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")
```

La participacin parental baja tiene una diferencia significativa en el puntaje comparado con la participacin parental elevada, la diferencia es significativa. Tambin hay una diferencia significativa entre la participacin parental baja y media. La diferencia entre la participacin parental elevada y media tambin es significativa. Entonces la participacin parental elevada tiene un impacto positivo y significativo en el rendimiento acadmico de los estudiantes, superando tanto la participacin media como baja.
