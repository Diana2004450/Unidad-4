>**Diseño experimental**

El diseño experimental es un enfoque metodológico que permite identificar relaciones causales entre variable

>**ANOVA**

Es una técnica estadística que se utiliza para determinar si existen diferencias significativas entre las medias de tres o más grupos. Este método analiza cómo una o más variables independientes afectan a una variable dependiente

>**Prueba de Tukey**

Es una técnica de comparación múltiple que se utiliza después de realizar un ANOVA, siempre que este indique que existen diferencias significativas entre las medias de los grupos.

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Diana2004450/Unidad-4/refs/heads/main/dataset%20(1).csv")
df.drop(columns=[], inplace=True)
df.dropna(inplace=True)
df
```

# **participación parental**

```python
df['participacion_parental'].unique()
```

```python
import scipy.stats as stats
nivel_de_significancia =0.05

#Realizar ANOVA
anova_result = stats.f_oneway(
      df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
      df[df['participacion_parental'] == 'medio']['puntaje_examen'],
      df[df['participacion_parental'] == 'elevado']['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos de los grupos.")
else:
    print("Conclusión: No hay diferencias significativas entre los grupos.")
```

El F-Estadístico de 80.46 indica que la variación entre las medias de los grupos, es decir, los niveles de participación parental es mucho mayor que la variación dentro de los grupos. El valor p = 0.00000 es mucho menor que el nivel de significancia de 0.05, entonces confirma que hay diferencias significativas entre las medias de al menos dos niveles de participación parental, obteniendo como conclusión que si existen diferencias significativas entre al menos dos de los grupos.

```python
# Si todos los grupos tienen una distribución normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los gráficos de Tukey para tomar una decisión (ignorando la tabla).
#.    Estrictamente hablando, tendrían que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

La participación parental baja tiene una diferencia significativa en el puntaje comparado con la participación parental elevada, la diferencia es significativa. También hay una diferencia significativa entre la participación parental baja y media. La diferencia entre la participación parental elevada y media también es significativa. Entonces la participación parental elevada tiene un impacto positivo y significativo en el rendimiento académico de los estudiantes, superando tanto la participación media como baja.
